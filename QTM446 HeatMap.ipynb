```python
import gmaps
gmaps.configure(api_key='AIzaSyApsnv-diA8hrDMrfEEp6Ci6tfUYN7POzg')


def drawHeatMap(locations, zoom, intensity, radius):
    # setting the data and parameters
    heatmap_layer = gmaps.heatmap_layer(locations, dissipating = True)
    heatmap_layer.max_intensity = intensity
    heatmap_layer.point_radius = radius
    # draw the heatmap into a figure
    fig = gmaps.figure()
    fig = gmaps.figure(center = [41.850033, -87.6500523], zoom_level=zoom)
    fig.add_layer(heatmap_layer)
    return fig
```
